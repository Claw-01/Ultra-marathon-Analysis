### Import necessary libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/TWO_CENTURIES_OF_UM_RACES.csv')
```

### See the data that's been imported

```python
df.head(10)
```

```python
df.dtypes
```

### Clean up data: Only want ESP Races, 2020

```python
df.value_counts('Event distance/length')
```

```python
df[df['Event distance/length'].isin(['50km'])]
```

```python
df2 = df[(df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'ESP')]
```

```python
df2.shape
```

```python
df2.head(10)
```

### Remove (ESP) from 'Event name'

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(10)
```

### Clean up athlete age

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth'].astype('Int64')
```

```python
df2.head(10)
```

### Remove text from 'Athlete performance'

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(10)
```

### Drop unnecessary columns

```python
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
df2.head()
```

### Clean up null values

```python
df2.isna().sum()
```

```python
df[df['Athlete year of birth'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

### Check duplicates

```python
df2.duplicated().sum()
```

### Reset index

```python
df2.reset_index(drop=True)
```

### Fix types

```python
df2.dtypes
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

### Rename columns

```python
df2 = df2.rename(columns = {'Year of event': 'year',
                            'Event dates': 'race_day',
                            'Event name': 'race_name',
                            'Event distance/length': 'race_length',
                            'Event number of finishers': 'race_num_finishers',
                            'Athlete performance': 'athlete_performance',
                            'Athlete gender': 'athlete_gender',
                            'Athlete average speed': 'athlete_average_speed',
                            'Athlete ID': 'athlete_id',

})
```

```python
df2.head()
```

```python
df2
```

### Analysis

```python
sns.histplot(df2['race_length'])
plt.xticks(rotation=45)
plt.show()
```

```python
sns.violinplot(data=df2, x='athlete_gender', y='athlete_average_speed', hue='athlete_gender', inner='quartile')
```

#### Filter: 50K and 53K

```python
df2_50k53k = df2[df2['race_length'].isin(['50km', '53km'])]
sns.violinplot(data=df2_50k53k, x='athlete_gender', y='athlete_average_speed', hue='athlete_gender', inner='quartile')
```

```python
df2_50k53k["age_group"] = pd.cut(df2_50k53k["athlete_age"], bins=[18, 25, 35, 45, 55, 65, 75, 100], labels=["18-25", "26-35", "36-45", "46-55", "56-65", "66-75", "76+"])

sns.violinplot(data=df2_50k53k, x="age_group", y="athlete_average_speed", inner="quartile")
plt.xticks(rotation=45) 
plt.show()
```

```python
df2_50k53k.dtypes
```

```python
sns.lmplot(data=df2_50k53k, x="athlete_age", y="athlete_average_speed", hue = 'athlete_gender')
```

```python
df2_50k53k.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
df2_50k53k.groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending=False).query('count>12')
```

```python

df2_unique_races = df2.drop_duplicates(subset=["race_name"])[["race_name", "race_num_finishers"]]


df2_popular_races = df2_unique_races[df2_unique_races["race_num_finishers"] > 200]


df2_popular_races = df2_popular_races.sort_values("race_num_finishers", ascending=False)


sns.barplot(
    data=df2_popular_races, 
    x="race_name", 
    y="race_num_finishers"
)
plt.xticks(rotation=90)
plt.title("Carreras con más de 200 finalistas")
plt.show()
```

```python
df2_gender_counts = df2.groupby(["race_name", "athlete_gender"])["athlete_id"].count().reset_index()

df2_gender_counts = df2_gender_counts.rename(columns={"athlete_id": "athlete_count"})

top_races = df2_gender_counts.groupby("race_name")["athlete_count"].sum().sort_values(ascending=False).index

sns.barplot(
    data=df2_gender_counts, 
    x="race_name", 
    y="athlete_count", 
    hue="athlete_gender",
    order=top_races
)
plt.xticks(rotation=90)
plt.title("Número de participantes por carrera y género")
plt.ylabel("Número de atletas")
plt.xlabel("Carrera")
plt.legend(title="Género")
plt.show()
```

```python
df2_unique_athletes = df2.drop_duplicates(subset=["athlete_id"])

sns.boxplot(data=df2_unique_athletes, x="race_name", y="athlete_age")
plt.xticks(rotation=90)
plt.title("Distribución de edades en cada carrera")
plt.show()
```

```python
df2['race_month'] = df2['race_day'].str.split('.').str.get(1).astype(int)
df2['race_season'] = df2['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring')
```

```python
df2_unique_races = df2.drop_duplicates(subset=["race_name", "race_day"])

sns.countplot(data=df2_unique_races, x="race_season", order=["Winter", "Spring", "Summer", "Fall"])
plt.title("Número de carreras por estación del año")
plt.show()
```
